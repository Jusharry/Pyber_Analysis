```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
#set the x-axis to a list of strings for each month
x_axis = ["Jan","Feb","Mar","April","May","June","July","Aug","Sept","Oct","Nov","Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis,y_axis, c = 'green',marker='D',label = 'Boston')
plt.xlabel('Date')
plt.ylabel('Fares($)')
plt.grid()
plt.title('PyBer Fare by Month')
plt.ylim(0, 45)
plt.legend()
plt.show()
```

```python
#create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
#create the plot
plt.bar(x_axis,y_axis,color='green',label='Boston')
#add labels
plt.xlabel('Date')
plt.ylabel('Fares($)')
#create a title
plt.title('PyBer Fare by Month')
#Add the legend
plt.legend()
```

```python
#Create a plot using horizontal axes with inversion 
plt.barh(x_axis,y_axis,color='magenta',label = 'Boston')
plt.ylabel('Date')
plt.xlabel('Fares($)')
plt.title('PyBer Fare by Month')
plt.legend()
#y_axis displayed in ascending order 
plt.gca().invert_yaxis()
```

```python
#Plot w/ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis,color='cyan',label='Chicago')
plt.xlabel('Fares($)')
plt.ylabel('Date')
plt.legend()
plt.gca().invert_yaxis()
```

```python
plt.plot(y_axis,x_axis,'o',c='red',label='Chicago')
plt.xlabel('Fares($)')
plt.ylabel('Date')
plt.xlim(0,50)
plt.ylabel('Fares($)')
plt.legend()
plt.gca().invert_yaxis()
```

```python
#Create bubble plot
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Larger bubbles
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#Using python list comprehension 
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#Using obj.oriented interface
plt.figure(figsize=(2,0))
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], facecolor='skyblue',edgecolor='k',label='Boston')
plt.ylabel('Date')
plt.xlabel('Fares($)')
plt.title('PyBer Fare by Month')
plt.xlim(0,50)
plt.legend()
plt.gca().invert_yaxis()
```

```python
#Pie charts
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
colors=["slateblue","magenta","lightblue","green","yellowgreen","greenyellow","yellow","orange","gold","indianred","tomato","mistyrose"]
plt.subplots(figsize=(8,8))
plt.pie(y_axis,explode = explode_values,colors=colors, labels = x_axis,autopct= '%.1f%%')
plt.show()
```

```python
explode = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
colors=["slateblue","magenta","lightblue","green","yellowgreen","greenyellow","yellow","orange","gold","indianred","tomato","mistyrose"]

fig, ax= plt.subplots(figsize=(8,8))
ax.pie(y_axis,labels=x_axis,explode= explode,colors=colors,autopct='%.1f%%',shadow=True,startangle= 95,counterclock=False)

plt.show()
```
