```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
battle_kings_df=pd.read_csv('Resources/got.csv')
battle_kings_df
```

```python
# Get attacker and defender king data
attacker_king = battle_kings_df['attacker_king'].value_counts()
attacker_king
```

```python
defender_king = battle_kings_df['defender_king'].value_counts()
defender_king
```

```python
# Get total battle data
total_battle_data= attacker_king.add(defender_king,fill_value=0)
total_battle_data
```

```python
# Configure plot and ticks
total_battle_data.plot(kind='bar',facecolor='red')

# Set textual properties
plt.title("Battle Royale")
plt.xlabel('Warriors')
plt.ylabel('Number of Battles')
plt.xticks(rotation=45)

# Show plot

plt.show()
# Resize plot to display labels

plt.tight_layout()
```
